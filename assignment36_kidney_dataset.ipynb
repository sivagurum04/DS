```python
import matplotlib.pyplot as plt
```

```python
cols = ['age','bp','sg','al','su','bgr','bu','sc','sod','pot','hemo']

plt.boxplot(df[cols])
plt.xticks(ticks=range(1, len(cols)+1), labels=cols, rotation=45)
plt.show()
```

```python
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[~((df[cols] < (Q1 - 1.5 * IQR)) | (df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Before:", df.shape)
print("After:", df_cleaned.shape)
```
