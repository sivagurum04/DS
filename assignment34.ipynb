```python
import matplotlib.pyplot as plt
x = [5, 7, 8, 7, 2]
y = [99, 86, 87, 88, 111]
plt.scatter(x, y)
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
x2 = [1, 2, 3, 4]
y2 = [40, 30, 20, 10]
plt.scatter(x1, y1, color='blue')
plt.scatter(x2, y2, color='red')
```

```python
x = [2, 4, 6, 8, 10]
y = [5, 10, 15, 20, 25]
sizes = [50, 100, 200, 300, 400]
plt.scatter(x, y, s=sizes)
plt.show()
```

```python
x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 35, 45]
colors = [1, 2, 3, 4, 5]
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.bar(categories, values)
plt.show()
```

```python
categories = ['Python', 'Java', 'C++', 'SQL']
values = [80, 65, 70, 75]
plt.barh(categories, values, color='green')
plt.show()
```

```python
data = [22, 87, 5, 43, 56, 73, 55, 54, 11, 20]
plt.hist(data, bins=5)
plt.title("Score Distribution")
plt.show()
```

```python
values = [35, 25, 25, 15]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
explode = [0.2, 0, 0, 0]
plt.pie(values, labels=labels, explode=explode,shadow=True)
plt.legend()
plt.show()
```
