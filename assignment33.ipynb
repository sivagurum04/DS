```python
import matplotlib
import matplotlib.pyplot as plt
print(matplotlib.__version__)
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()
```

```python
y = [5, 15, 10, 25, 20]
plt.plot(y)
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [3, 7, 1, 9]
plt.plot(x, y, marker='*', linestyle='None')
plt.show()
```

```python
y = [3, 8, 1, 10]
plt.plot(y, marker='o', linestyle=':', color='r', markersize=10)
plt.show()
```

```python
y1 = [1, 3, 5, 7]
y2 = [2, 4, 6, 8]
plt.plot(y1, color='blue')
plt.plot(y2, color='green')
plt.show()
```

```python
x = [10, 20, 30, 40]
y = [15, 25, 35, 45]
plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed vs Time")
```

```python
y = [4, 6, 2, 8, 10]
plt.plot(y, linestyle='--', linewidth=5, color='green')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([3, 8, 1, 10])
plt.title("Plot 1")
plt.subplot(1, 2, 2)
plt.plot([5, 2, 7, 11])
plt.title("Plot 2")
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot([1, 4, 9, 16])
plt.title("Squares")
plt.subplot(2, 1, 2)
plt.plot([2, 8, 18, 32])
plt.title("Multiples")
plt.suptitle("My Graphs")
plt.show()
```
