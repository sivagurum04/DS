```python
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
outlier=[]
for i in df["Age"]:
    if (i>upper_limit) or (i<lower_limit):
        outlier.append(i)
```
