```python
import pandas as pd
sales_data = {
    'Store': ['Chennai','Chennai','Bangalore','Bangalore','Hyderabad',
              'Hyderabad','Chennai','Bangalore','Hyderabad','Chennai'],
    'Category': ['Electronics','Clothing','Electronics','Groceries','Clothing',
                 'Groceries','Groceries','Clothing','Electronics','Electronics'],
    'Year': [2023,2023,2023,2023,2023,2024,2024,2024,2024,2024],
    'Sales': [25000,18000,22000,15000,16000,21000,19000,17000,23000,26000]
}

df_sales = pd.DataFrame(sales_data)
```

```python
df_sales.groupby('Store')['Sales'].sum()
```

```python
df_sales.groupby('Category')['Sales'].mean()
```

```python
df_sales.groupby('Year')['Sales'].sum()
```

```python
df_sales[df_sales['Year'] == 2024]
```

```python
df_sales.groupby('Store').get_group('Chennai')
```

```python
students = pd.DataFrame({
    'student_id':[1,2,3,4,5],
    'Name':['Arun','Divya','Karthik','Meena','Rahul'],
    'course_id':['C1','C2','C3','C1','C2']
})

courses = pd.DataFrame({
    'course_id':['C1','C2','C3','C4'],
    'Course_Name':['Python','Data Science','Power BI','AI']
})
```

```python
pd.merge(students, courses, on='course_id', how='inner')
```

```python
pd.merge(students, courses, on='course_id', how='left')
```

```python
outer_df = pd.merge(students, courses, on='course_id', how='outer')
print(outer_df)

# Check missing values
outer_df.isnull()
```
